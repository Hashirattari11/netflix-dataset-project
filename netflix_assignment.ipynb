```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("netflix_titles.csv")

df
```

```python
df.head()
```

```python
null = df.isnull().sum()
null
```

```python
print(df.info())
```

```python
print(df.columns)
```

```python
print(df["type"].value_counts())
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x="type" , data=df )
plt.title("Movies vs Tv show")
plt.show()
```

```python
topcountry = df['country'].value_counts().head(10)
topcountry
```

```python
plt.figure(figsize=(8,6))
topcountry.plot(kind="barh" , color= "skyblue")
plt.title("The top 10 country by netflix content ")
plt.show()
```

```python
df['Main_Genre'] = df['listed_in'].str.split(',').str[0]
print("\nMost Common Genres:\n", df['Main_Genre'].value_counts().head(10))
```

```python
df['rating'].value_counts().head(6).plot(kind='pie', autopct='%1.1f%%')
plt.title("Content Ratings Distribution")
plt.ylabel("")
plt.show()
```

```python
print("\n Top highest number release" , df["release_year"].value_counts().head(10))
```

```python
df['release_year'].value_counts().sort_index().plot(kind="line")
plt.title("Number of release year")
plt.show()
```

```python
print("netflix rating" ,df['rating'].value_counts())
```

```python
pivot = pd.crosstab(df['country'], df['Main_Genre'])
sns.heatmap(pivot.head(10), cmap="YlGnBu")
plt.title("Country vs Genre Heatmap (Top 10 Countries)")
plt.show()
```

```python
top_directors = df['director'].value_counts().head(10)
top_directors
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_directors.values, y=top_directors.index, palette="coolwarm")
plt.title("Top 10 Directors by Number of Titles on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Director")
plt.show()
```
